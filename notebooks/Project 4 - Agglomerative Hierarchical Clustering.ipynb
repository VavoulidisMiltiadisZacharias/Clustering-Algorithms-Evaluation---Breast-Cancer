###  Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import random
import time

from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
```

### Suppress warnings

```python
# Suppress all warnings
import warnings

warnings.filterwarnings("ignore")
```

### Start time

```python
# Start time
start_time = time.time()
```

### Load the final dataset and store it within a pandas DataFrame

```python
# Load the final dataset and store it withina  panda dataframe to handle it easier
df = pd.read_csv(r"C:\Users\user\Desktop\VAVOULIDIS MILTIADIS - ZACHARIAS\Projects\Project 4\final_dataset.csv")

# Print the dataset
df
```

### Split features and true label

```python
X = df.iloc[:, :-1]   # all columns except last
y = df.iloc[:, -1]    # last column ("Class")
```

### Apply K-Means

```python
### Apply Agglomerative Hierarchical Clustering
agglo = AgglomerativeClustering(n_clusters=2, linkage='ward')
clusters_agglo = agglo.fit_predict(X)
```

### Evaluate clustering (scientific way)

```python
ari_agglo = adjusted_rand_score(y, clusters_agglo)
nmi_agglo = normalized_mutual_info_score(y, clusters_agglo)

print("Agglomerative ARI:", ari_agglo)
print("Agglomerative NMI:", nmi_agglo)
```

### End time

```python
# End time
end_time = time.time()
```

### Elapsed time

```python
# Elapsed time
elapsed_time = end_time - start_time

print(f"Elapsed time for Agglomerative Hierarachical Clustering: {elapsed_time:.2f} seconds")
```
